```python
# 7458 chip
%%writefile stepone.v
module top_module (
    input p1a, p1b, p1c, p1d, p1e, p1f,
    output p1y,
    input p2a, p2b, p2c, p2d,
    output p2y );

    assign p1y = (p1a & p1b & p1c) | (p1d & p1e & p1f);
    assign p2y = (p2a & p2b) | (p2c & p2d);
endmodule
```

```python
%%writefile tb.v
`timescale 1ns/1ns

module tb;
    wire p1y_tb, p2y_tb;
    reg p1a_tb, p1b_tb, p1c_tb, p1d_tb, p1e_tb, p1f_tb, p2a_tb, p2b_tb, p2c_tb, p2d_tb;

    // Instantiate the stepone module
    top_module uut (
        .p1a(p1a_tb), .p1b(p1b_tb), .p1c(p1c_tb), .p1d(p1d_tb), .p1e(p1e_tb), .p1f(p1f_tb), .p2a(p2a_tb), .p2b(p2b_tb), .p2c(p2c_tb), .p2d(p2d_tb), .p1y(p1y_tb), .p2y(p2y_tb)
    );

    initial begin
        $dumpfile("tb.vcd");  // Make sure we generate tb.vcd
        $dumpvars(0, tb);     // Dump all signals in tb module

        # 0 p1a_tb = 0; p1b_tb = 1; p1c_tb = 1; p1d_tb = 1; p1e_tb =1; p1f_tb = 1; p2a_tb = 0; p2b_tb =1; p2c_tb =1; p2d_tb =1;
        # 5 p1a_tb = 0; p1b_tb = 0; p1c_tb = 1; p1d_tb = 1; p1e_tb =1; p1f_tb = 0; p2a_tb = 0; p2b_tb =1; p2c_tb =1; p2d_tb =1;
        # 5 p1a_tb = 0; p1b_tb = 1; p1c_tb = 1; p1d_tb = 0; p1e_tb =0; p1f_tb = 1; p2a_tb = 0; p2b_tb =0; p2c_tb =0; p2d_tb =1;
        # 5 p1a_tb = 0; p1b_tb = 0; p1c_tb = 0; p1d_tb = 1; p1e_tb =1; p1f_tb = 1; p2a_tb = 0; p2b_tb =1; p2c_tb =1; p2d_tb =1;

        // Monitor time and output signal
        $monitor("Time = %0t ns | Output_p1y = %b | Output_p2y = %b", $time, p1y_tb, p2y_tb);

        #10 $finish;  // Run simulation for 10 time units
    end
endmodule
```
